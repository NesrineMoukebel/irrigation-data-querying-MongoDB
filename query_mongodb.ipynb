```python
from pymongo import MongoClient

# Connexion à MongoDB
client = MongoClient('localhost', 27017)
db = client['agriculture']
collection = db['irrigation_data']
```

```python
# QUERY 1 
# Get all documents
print("Get all documents :")
for document in collection.find():
    print(document)
```

```python
# QUERY 2
# Get documents where Irrigation is necessary
print("\nDocuments where Irrigation is necessary :")
for document in collection.find({"Irrigation(Y/N)": 1}):
    print(document)
```

```python
# QUERY 3 
print("\nDocuments where soil temperature > 19°C :")
for document in collection.find({"Soil Temperature": {"$gt": 19}}):
    print(document)
```

```python
# QUERY 4
print("\nDocuments with humidity > 80% :")
for document in collection.find({"Humidity": {"$gt": 80}}):
    print(document)
```

```python
# QUERY 5
print("\nDocuments for Type2 Crop where irrigation is necessary:")
for document in collection.find({"CropType": 2, "Irrigation(Y/N)": 1}):
    print(document)
```

```python
# QUERY 6
print("\nDocuments where crop days are more than 90 jours :")
for document in collection.find({"CropDays": {"$gt": 90}}):
    print(document)
```

```python
# QUERY 7
print("\nDocuments with temperature between 20 and 30°C :")
for document in collection.find({"Temperature": {"$gte": 20, "$lte": 30}}):
    print(document)
```

```python
# QUERY 8
print("\nDocuments with temperature > 30°C or soil humidity < 40% :")
for document in collection.find({"$or": [{"Temperature": {"$gt": 30}}, {"Soil Moisture": {"$lt": 40}}]}):
    print(document)
```
